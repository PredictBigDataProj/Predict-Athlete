# One-hot encode preferrred foot

```python
import pandas as pd
```

```python
df_players = pd.read_csv('Finished_final_proj.csv')  # index=False avoids saving row indices
```

```python
df_players['preferred_foot'].head()
```

```python
unique_foot = list(df_players['preferred_foot'].unique())
unique_foot
```

```python
def one_hot_encode_foot(df_players, unique_foot):
    df_players_foot_encoded = pd.DataFrame()
    for foot in unique_foot:
        df_players_foot_encoded[f'preferred_foot_{foot}'] = df_players['preferred_foot'].apply(lambda x: foot_present(foot, x))

    df_players = pd.concat([df_players, df_players_foot_encoded], axis=1)
    return df_players



def foot_present(foot, unique_foot):
    if foot in unique_foot:
        return 1
    else:
        return 0

df_players = one_hot_encode_foot(df_players, unique_foot)

#print(df.head(5))
```
