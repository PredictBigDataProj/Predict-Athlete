# Get starting year for players in their career

```python
df['Season'].head()
```

```python
def get_start_year(season_years_list):

    years_list = ast.literal_eval(season_years_list)
        
        


    num_length = len(years_list)

    start_year =  years_list[num_length-1]

    return start_year


df['Start_year'] = df['Season'].apply(get_start_year)

df[['Season', 'Start_year']].head()
```

```python
def get_recent_end_year(season_years_list):

    years_list = ast.literal_eval(season_years_list)
        
        


    #num_length = len(years_list)

    end_year =  years_list[0]

    return end_year


df['End/Recent_year'] = df['Season'].apply(get_recent_end_year)

df[['Season', 'End/Recent_year']].head()
```

```python
df.info()
```

```python
df.head()
```

```python
df['Career_length'] = 1
```

```python
df.info()
```

```python


def extract_start_year(start_year_string):

    if '/' in start_year_string:
        year = str(start_year_string.split('/')[0])
        return year
    else:
        return start_year_string


df['Start_year_final'] = df['Start_year'].apply(extract_start_year)

df[['Start_year', 'Start_year_final']].head()
```

```python
def extract_end_year(start_year_string):

    if '/' in start_year_string:
        year = str(start_year_string.split('/')[0])

        return year
    else:
        return start_year_string


df['End/Recent_year_final'] = df['End/Recent_year'].apply(extract_end_year)

df[['End/Recent_year', 'End/Recent_year_final']].head()
```

```python
year_map = {
    'empty': '0'
}

df['Start_year_final'] = df['Start_year_final'].replace(year_map)
df['End/Recent_year_final'] = df['End/Recent_year_final'].replace(year_map)
```

```python
df['Start_year_final'] = df['Start_year_final'].astype(int)
df['End/Recent_year_final'] = df['End/Recent_year_final'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Career_length'] = df['End/Recent_year_final'] - df['Start_year_final']
```

```python
df.info()
```

```python
df.head()
```

```python
career_year_map = {
    0: 1
}

df['Career_length'] = df['Career_length'].replace(career_year_map)
```

```python
df.to_csv('sofifa-career-length.csv', encoding='utf-8-sig')
```
