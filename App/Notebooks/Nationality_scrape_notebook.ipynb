```python
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import random
from bs4 import BeautifulSoup


def extract_season_club_data(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    tables = soup.find_all('table')

    season_club_data = []


    if not tables:
        season_club_data.append(("empty", "empty"))
    else:
        rows = tables[0].find_all('tr')[1:]  
        for row in rows:
            cells = row.find_all('td')
            if len(cells) > 1:
                season = cells[0].text.strip()
                club = cells[1].find('a').text.strip()
                season_club_data.append((season, club))
    
    # for table in tables:
    #     rows = table.find_all('tr')[1:]  # Skip the header row
    #     for row in rows:
    #         cells = row.find_all('td')
    #         if len(cells) > 1:
    #             season = cells[0].text.strip()
    #             club = cells[1].find('a').text.strip()
    #             season_club_data.append((season, club))

    return season_club_data


def extract_national_season_nation_data(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    tables = soup.find_all('table')


    national_season_nation_data = []

    if not tables:
        national_season_nation_data.append(("empty", "empty"))
    else:
        rows = tables[1].find_all('tr')[1:] 
        for row in rows:
            cells = row.find_all('td')
            if len(cells) > 1:
                national_season = cells[0].text.strip()
                nation = cells[1].find('a').text.strip()
                national_season_nation_data.append((national_season, nation))
    
    # for table in tables:
    #     rows = table.find_all('tr')[1:]  # Skip the header row
    #     for row in rows:
    #         cells = row.find_all('td')
    #         if len(cells) > 1:
    #             season = cells[0].text.strip()
    #             club = cells[1].find('a').text.strip()
    #             season_club_data.append((season, club))

    return national_season_nation_data


with open('player-urls-full.csv', newline='', encoding='utf-8-sig') as csvfile:
    reader = csv.reader(csvfile)
    urls = [row[0] for row in reader]


driver = webdriver.Chrome()


with open('sofifa-scrape-season-12.csv', mode='w', newline='', encoding='utf-8-sig') as csvfile:
    writer = csv.writer(csvfile)

    writer.writerow(['URL', 'Player Name', 'Season', 'Club/Nation', 'Type'])

    for url in urls:
        print(f"Processing URL: {url}")

      
        try:
            player_name = url.split('/player/')[1]  
        except IndexError:
            player_name = "Unknown"

      
        driver.set_page_load_timeout(300)
        
        driver.get(url)
        rand = random.randint(5, 10) 
        time.sleep(rand)
       
        season_club_data = extract_season_club_data(driver)
        national_season_nation_data = extract_national_season_nation_data(driver)

       
        for season, club in season_club_data:
            writer.writerow([url, player_name, season, club, 'League'])


        for national_season, nation in national_season_nation_data:
            writer.writerow([url, player_name, national_season, nation, 'National'])


driver.quit()

print("Scraping complete! Results saved to 'sofifa-scrape-season-11.csv'")
```

```python
import pandas as pd


players = pd.read_csv("semi_clean_sofifa_players.csv")
nation = pd.read_csv("nation-careers.csv")
league = pd.read_csv("league-careers.csv")


nation_merged = players.merge(
    nation, how="left", left_on="normalized_name", right_on="Name")


nation_merged_alt = players.merge(
    nation, how="left", left_on="normalized_full_name", right_on="Name")

nation_merged = nation_merged.combine_first(
    nation_merged_alt[nation_merged_alt["Name"].notna()])


nation_merged.dropna(subset=["Name"], inplace=True)
nation_merged.to_csv("nation_merged.csv", index=False)


league_merged = players.merge(
    league, how="left", left_on="normalized_name", right_on="Name")


league_merged_alt = players.merge(
    league, how="left", left_on="normalized_full_name", right_on="Name")

league_merged = league_merged.combine_first(
    league_merged_alt[league_merged_alt["Name"].notna()])

league_merged.dropna(subset=["Name"], inplace=True)
league_merged.to_csv("league_merged.csv", index=False)



nation_merged = pd.read_csv("nation_merged.csv")
nation_careers = pd.read_csv("nation-careers.csv")

merged_ids = set(nation_merged["player_id"])


career_ids = set(
    nation_merged[nation_merged["normalized_name"].isin(nation_careers["Name"])]["player_id"])


missing_ids = merged_ids - career_ids

print("Player IDs in nation_merged.csv but not in nation-careers.csv:")
print(missing_ids)


missing_players = nation_merged[nation_merged["player_id"].isin(missing_ids)][["player_id", "name"]]
print("\nMissing players:")
print(missing_players)
```
